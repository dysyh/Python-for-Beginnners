- We should consider which of the three output sentences is most important? In this example, we want the thrid to be the most important one.
- We can rewrite the program below

```python
age = 10

if 5<age<20:
    print('your age younger than 20 but older than 5')
elif age<5:
    print('age younger than 5')
else:
    print('age is older than 20')
```

# 5 - for loop

- for loop will print everything inside the variable, which is a list of numbers in this exmaple
- In the small program below, the x is just the index to count each number in the num list from 0, you can use any name to replace

```python
num = [1,2,3,4,5,6,7]

for x in num:
    print(x)
```

```python
for i in num:
    print((i))
```

- "x" and "i" are doing same thing. Just to count the index inside num list from 0.

- Another example below:

```python
name = ['Alex','Bob','Jason','Micheal']

for i in name:
    print(i)
```

 - Explaination of the code above: 
 - First, i will count the values inside name list from 0 to 3 becuase there are 4 values in this list.
 - Second, it will do as in section 3 above. When i is 0, name[i] is name[0], which will print out "Alex"; 
 - When i is 1, name[i] is name[1], which will print out "Bob".
 - The process will continue until every single value inside the list printed out

```python
print(len(name))  #len() is a function to calculate how many values in one variable. In this example, it has 4 names
```

# 6 - range

 - When we want to print out numbers without crating a list, we can use range

```python
for i in range(5):
    print(i)
```

 - The above program print out values from 0 to 5.
 - It has 5 values that ranges from 0 to 4; Becuase Python counts from 0 so the value 5 doesn't dispaly.

```python
for i in range(1,5):
    print(i)
```

 - The above program sets the range starts from 1, end at 5, so the output has 4 values will be from 1 to 4.

```python
for i in range(1,5,2):
    print(i)
```

 - The above program sets the range from 1, end at 4, with increments of 2 each time. 
 - So its like: 1,3,5,7,9,11,13,15. But we set the maximum number can be output is 4, so the output is only 1 and 3. 

# 7 - While

 - "While" means "while something is true, execute the program by print out something"

```python
Age = 20

while age<30:
    print('age is',Age)
```

- In the above program, "while age<30" is true, so print out "age is 20". However, it will contine forever until your computer craches.
- We can add a limit to it as in the exmaple below, so the program will stop at a certain point

```python
Age = 20

while age<40:
    print('age is',Age)
    Age=Age+1            #Can also write as Age+=1
```

- The program will stop when at age is 39 becuase its smaller than 40. 
- The program starts from age=20, which is smaller than 40, so it will print out "age is 20", then Age=Age+1 became 21;
- 21 is still smaller than 40, it print out "age is 21", ten Age=Age+1 became 22;
- The process contines until when age=39, as usual, it print out "age is 39", then Age=Age+1 became 40.
- Age is now 40. while age<40 is not true anymore becuase 40 is not smaller than 40. Program will stop here.

```python
def apple(eat_apple):
    
    life = 3*eat_apple
    print('your life will increase',life,'days')
    
apple(10)              #Call apple to eat 10 apples
    
```
